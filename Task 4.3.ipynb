```python
img1 = cv2.imread("tony.jpg")
img2 = cv2.imread("captian.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```
