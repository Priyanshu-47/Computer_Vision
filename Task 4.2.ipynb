```python
import cv2
import numpy as np
import matplotlib.pyplot as plot
```

```python
pic1 = cv2.imread("c.jpg")
pic2 = cv2.imread("b.jpg")
pic1 = cv2.cvtColor(pic1, cv2.COLOR_BGR2RGB)
pic2 = cv2.cvtColor(pic2, cv2.COLOR_BGR2RGB)
```

```python
plot.imshow(pic1)
```

```python
plot.imshow(pic2)
```

```python
pic1_arr = pic1.copy()
pic2_arr = pic2.copy()
```

```python
pic1_arr_cropped = pic1_arr[50:600, 400:1000]
pic1_arr_cropped.shape
```

```python
plot.imshow(pic1_arr_cropped)
cv2.imwrite("image1.jpg",pic1_arr_cropped)
```

```python
pic2_arr_cropped= pic2_arr[50:600, 150:750]
pic2_arr_cropped.shape
```

```python
plot.imshow(pic2_arr_cropped)
cv2.imwrite("image1.jpg",pic2_arr_cropped)
```

```python
pic1[50:600, 150:750] = pic2_arr_cropped
plot.imshow(pic1)
```

```python
pic2[50:600, 100:700] = pic1_arr_cropped
plot.imshow(pic2)
```
